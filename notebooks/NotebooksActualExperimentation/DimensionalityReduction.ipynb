This notebook contains the necessary code for the dimensionality reduction operations. The dimensionality reduction is performed to select the label based on how separable the class is to the rest of labels.

### Dimensionality reduction with the Human Activity Recognition (HAR) dataset

```python
dataset_factory = dataset_model_dictionary["har"]()
X_train, y_train = dataset_factory.get_dataset().get_training_data()
X_test, y_test = dataset_factory.get_dataset().get_test_data()
```

```python
y_train_compressed = np.argmax(y_train, axis=1)
print(y_train_compressed)
```

```python
import plotly.express as px
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

Dimensionality reduction using Principal Component Analysis (PCA)

```python
pca = PCA(n_components=2)
components = pca.fit_transform(X_train)

fig = px.scatter(components, x=0, y=1, color=y_train_compressed)
fig.show()
```

```python
pca = PCA(n_components=3)
components = pca.fit_transform(X_train)

fig = px.scatter_3d(components, x=0, y=1, z=2, color=y_train_compressed)
fig.show()
```

Dimensionality reduction using Linear Discriminant Analysis (LDA)

```python
lda = LDA(n_components=2)
# Taking in as second argument the Target as labels
components = lda.fit_transform(X_train, y_train_compressed)

fig = px.scatter(components, x=0, y=1, color=y_train_compressed)
fig.show()
```

```python
lda = LDA(n_components=3)
# Taking in as second argument the Target as labels
components = lda.fit_transform(X_train, y_train_compressed)

fig = px.scatter_3d(components, x=0, y=1, z=2, color=y_train_compressed)
fig.show()
```

Dimensionality reduction using T-Distributed Stochastic Neighbor Embedding (TSNE)

```python
tsne = TSNE(n_components=2)
# Taking in as second argument the Target as labels
components = tsne.fit_transform(X_train)

fig = px.scatter(components, x=0, y=1, color=y_train_compressed)
fig.show()
```

```python
tsne = TSNE(n_components=3)
# Taking in as second argument the Target as labels
components = tsne.fit_transform(X_train)

fig = px.scatter_3d(components, x=0, y=1, z=2, color=y_train_compressed)
fig.show()
```
